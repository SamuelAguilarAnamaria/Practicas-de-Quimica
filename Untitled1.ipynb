```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros generales
n = 1       # Factor de idealidad (aproximadamente 1 para diodos de silicio)
VT = 26e-3  # Tensión térmica (V)

def diodo_ideal(V):
    Is = 1e-12  # Corriente de saturación inversa (A)
    I = Is * (np.exp(V / (n * VT)) - 1)
    return I

def diodo_zener(V):
    Vz = 10  # Tensión Zener (V)
    Iz = 1e-3  # Corriente Zener (A)
    I = Iz * (np.exp(V / (n * VT)) - 1)
    return I

def diodo_tunnel(V):
    Is = 1e-12  # Corriente de saturación inversa (A)
    I = Is * (np.exp(V / (n * VT)) - 1)
    return I

def diodo_schottky(V):
    Is = 1e-12  # Corriente de saturación inversa (A)
    I = Is * (np.exp(V / (n * VT)) - 1)
    return I

def diodo_varicap(V):
    Cj = 10e-12  # Capacitancia del diodo sin polarización (F)
    Vb = 0.7     # Tensión de barrera en la unión (V)
    I = Cj / np.sqrt(1 - (V / Vb) ** 2)
    return I

def diodo_varistor(V):
    Cj = 10e-12  # Capacitancia del varistor sin polarización (F)
    Vb = 0.7     # Tensión de barrera en la unión (V)
    I = Cj / np.sqrt(1 - (V / Vb) ** 2)
    return I

def graficar_curva(diodo):
    V = np.linspace(-1, 1, 100)  # Rango de voltajes
    I = diodo(V)  # Calcula la corriente según el tipo de diodo

    plt.figure(figsize=(8, 6))
    plt.plot(V, I, label='Curva V-I')
    plt.xlabel('Voltaje (V)')
    plt.ylabel('Corriente (A)')
    plt.title(f'Curva V-I de un Diodo {diodo.__name__.capitalize()}')
    plt.grid(True)
    plt.legend()
    plt.show()

def main():
    print("Selecciona un tipo de diodo:")
    print("1. Ideal")
    print("2. Zener")
    print("3. Túnel")
    print("4. Schottky")
    print("5. Varicap")
    print("6. Varistor")

    # Define la opción aquí en lugar de usar input()
    opcion = 6

    if opcion == 1:
        graficar_curva(diodo_ideal)
    elif opcion == 2:
        graficar_curva(diodo_zener)
    elif opcion == 3:
        graficar_curva(diodo_tunnel)
    elif opcion == 4:
        graficar_curva(diodo_schottky)
    elif opcion == 5:
        graficar_curva(diodo_varicap)
    elif opcion == 6:
        graficar_curva(diodo_varistor)
    else:
        print("Opción no válida. Introduce un número del 1 al 6.")

if __name__ == "__main__":
    main()
```
